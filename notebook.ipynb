```python
#Splitting the data in to X and Y
y = df[["Churn"]]
x = df[["Age","Gender","Tenure","MonthlyCharges"]]
```

```python
x
```

```python
y
```

```python
 x["Gender"]= x["Gender"].apply(lambda x: 1 if x== "Female" else 0) # 1 = female and  for male = 0
```

```python
y["Churn"]
```

```python
y["Churn"] = y["Churn"].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train) 
```

```python
import joblib\
    
joblib.dump(scaler,"scaler.pkl")
```

```python
x_test = scaler.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
def model_performance(predictions):
    print(f"Accuracy score on model is: {accuracy_score(y_test,predictions)}")
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
log_model.fit(x_train,y_train)
```

```python
x
```

```python
y_pred = log_model.predict(x_test)
```

```python
model_performance(y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
param_grid = {"n_neighbors": [3, 5, 7, 9], "weights": ["uniform", "distance"]}
```

```python
gridkn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
```

```python
gridkn.fit(x_train,y_train)
```

```python
gridkn.best_params_
```

```python
y_pred = gridkn.predict(x_test)
```

```python
y_pred
```

```python
model_performance(y_pred)
```

```python
from sklearn.svm import SVC
```

```python
SVM = SVC()
```

```python
param_grid  = {"C": [0.01,0.1, 0.5, 1], "kernel": ["linear", "rbf","ploy"]}
```

```python
gridsvc = GridSearchCV(SVM, param_grid, cv=5)
```

```python
gridsvc.fit(x_train,y_train)
```

```python
gridsvc.best_params_
```

```python
y_pred = gridsvc.predict(x_test)
```

```python
model_performance(y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid = {"criterion": ["gini", "entropy"], 
              "splitter": ["best", "random"],
              "max_depth": [10,20,30, None],
              "min_samples_split": [2, 5, 10],
              "min_samples_leaf": [1, 2, 4]
              }
```

```python
grid_tree = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)
```

```python
y_pred = grid_tree.predict(x_test)
```

```python
model_performance(y_pred)
```
